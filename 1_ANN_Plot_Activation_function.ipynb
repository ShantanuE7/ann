```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Activation Functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    exp_z = np.exp(x)
    sum = exp_z.sum()
    softmax_z = np.round(exp_z/sum,len(x))
    return softmax_z
```

```python
x = np.linspace(-5, 5, 100)
```

```python
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)
y_softmax = softmax(x)
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y_sigmoid, label='Sigmoid')
plt.title('Sigmoid Activation Function')
plt.legend()
```

```python
plt.subplot(2, 2, 2)
plt.plot(x, y_relu, label='ReLU')
plt.title('ReLU Activation Function')
plt.legend()
```

```python
plt.subplot(2, 2, 3)
plt.plot(x, y_tanh, label='Tanh')
plt.title('Tanh Activation Function')
plt.legend()
```

```python
plt.subplot(2, 2, 4)
plt.plot(x, y_softmax, label='Softmax')
plt.title('Softmax Activation Function')
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```
